```python
import re
import pandas as pd
import numpy as np
```

```python
series=pd.read_csv('Data/tv_shows.csv')
```

```python
#Unnamed: 0 is the same as index
(series.index==series['Unnamed: 0']).sum()
```

```python
#type id being kept in case we want to merge series and movies
series.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
series.rename(columns={'type':'Type'},inplace=True)
```

```python
movies=pd.read_csv('Data/movies.csv')
```

```python
#Unnamed: 0 is the same as index
(movies.index==movies['Unnamed: 0']).sum()
```

```python
#ID the same as index+1
(movies.index+1==movies['ID']).sum()
```

```python
movies.drop(['Unnamed: 0','ID'],axis=1,inplace=True)
```

```python
movies.info()
```

```python
series.head()
```

```python
movies['Episodes']=1
```

```python
movies.head()
```

```python
extra_series=pd.read_csv('Data/extra_tv_shows.csv')
```

```python
full_series=pd.merge(series,extra_series)
```

```python
full_series.rename(columns={'Genre':'Genres','Creator':'Directors'},inplace=True)
```

```python
full_series=full_series[movies.columns]
```

```python
full_series.head()
```

```python
data=pd.concat([movies,full_series],ignore_index=True)
```

```python
data.info()
```

```python
data.iloc[16740:16750]
```

```python
data.rename({'Directors':'Directors/Creators','Language':'Languages','Country':'Countries'},axis=1,inplace=True)
```

```python
data['Rotten Tomatoes']=data['Rotten Tomatoes'].apply(lambda x:x if type(x)!=str else x.strip('%'))
data['Rotten Tomatoes']=data['Rotten Tomatoes'].astype(np.float64)
```

```python
data.Age=data.Age.apply(lambda x:x if type(x)!=str else 0 if re.findall('all',x) else x.strip('+'))
data.Age=data.Age.astype(np.float64)
```

```python
data.Runtime=data.Runtime.apply(lambda x:x if type(x)!=str else x.replace(',',''))
data.Runtime=data.Runtime.astype(np.float64)
```

```python
data.info()
```

```python
genres=list(set([y for x in data['Genres'].unique() for y in str(x).split(',')]))
directors=list(set([y for x in data['Directors/Creators'].unique() for y in str(x).split(',')]))
countries=list(set([y for x in data['Countries'].unique() for y in str(x).split(',')]))
languages=list(set([y for x in data['Languages'].unique() for y in str(x).split(',')]))
```

```python
data.to_csv('Data/full.csv',index=False,)
```

```python
new_data=data.copy()
multicolumns=['Directors/Creators','Genres','Countries','Languages']
```

```python
#Number of new rows for each index
rows=[]
for i in new_data.index:
    crea=str(new_data.loc[i,'Directors/Creators']).split(',')
    gen=str(new_data.loc[i,'Genres']).split(',')
    con=str(new_data.loc[i,'Countries']).split(',')
    lang=str(new_data.loc[i,'Languages']).split(',')
    row=new_data.copy().loc[i]
    for cr in crea:
        for ge in gen:
            for co in con:
                for la in lang:
                    row.loc[multicolumns[0]]=cr
                    row.loc[multicolumns[1]]=ge
                    row.loc[multicolumns[2]]=co
                    row.loc[multicolumns[3]]=la
                    rows.append(row.copy())
    if i%(int(len(new_data)/100))==0:
        print(f'Done index: {i}')
```

```python
megadata=pd.DataFrame(rows)
megadata.to_csv('Data/extra_full.csv',index=False,)
```

```python
seperated_genres = data['Genres'].str.get_dummies(',')
print('Done Genres')
seperated_genres.columns=['genre_'+x for x in seperated_genres.columns]
seperated_countries = data['Countries'].str.get_dummies(',')
print('Done Countries')
seperated_countries.columns=['countrie_'+x for x in seperated_countries.columns]
seperated_languages = data['Languages'].str.get_dummies(',')
print('Done Languages')
seperated_languages.columns=['language_'+x for x in seperated_languages.columns]
```

```python
data=pd.concat([data,seperated_genres,seperated_countries,seperated_languages],axis = 1, sort = False)
```

```python
data.to_csv('Data/dummies.csv',index=False)
```
